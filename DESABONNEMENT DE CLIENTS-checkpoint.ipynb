```python
# chargement du datasabset
valeurs_manquantes = ['Unknown']
data = pd.read_csv('DEPART_CLIENT.csv', na_values = valeurs_manquantes, sep = ';')
data.head()
```

```python
# Relation de la variable cible avec les variables qualitatives
data_quali = data[['ACTIVITE_CLIENT', 'SEXE', 'NIVEAU_ETUDE', 'STATUT_MATRI',
       'CATEGO_REVENU_ANNUEL', 'TYPE_DE_CARTE', 'NBRE_PERS_A_CHARGE', 'NBRE_MOIS_INACTIV', 'NBRE_MOIS_INACTIV_12', 'NBRE_TOTAL_PRODUITS_CLIENT']]
fig, axes = plt.subplots(5,2,figsize = (15,15)) 
for idx, cat_col in enumerate(data_quali):
    ligne, col = idx//2, idx%2
    sns.countplot(x = cat_col, data = data_quali, hue = 'ACTIVITE_CLIENT', ax = axes[ligne, col])
    plt.subplots_adjust(hspace = 1)
```

```python
data_quanti = data[['AGE',
'PERIOD_RELATIO_BANQUE',
'LIMITE_CREDIT',
'SOLDE_RENOUVELABLE',
'LIGNE_CREDIT_ACHAT',
'CHANGEMENT_MONTANT',
'TOTAL_TRANSAC',
'NBRE_TOTAL_TRANSAC',
'CHANG_NBRE_TRANSAC',
'TAUX_UTILI_MOYEN']]
```

```python
# Normalisation
normali = StandardScaler()
```

```python
data_normali = pd.DataFrame(normali.fit_transform(data_quanti), columns = data_quanti.columns)
```

```python
data_normali
```

```python
def plot_quanti(variable):
    plt.figure(figsize = (10,5))
    sns.boxplot(data['ACTIVITE_CLIENT'], data_quanti[variable])
    plt.suptitle('Distribution de ' + str(variable))
    return plt.show()
```

```python
interact(plot_quanti, variable = data_quanti.columns);
```
